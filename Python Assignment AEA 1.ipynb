# Motivation

One of the biggest debates in the current internationally and globally organized world has to do with taxation. More specifically, it is all about tax evasion. The published papers about tax evasion contributed a lot to this discussion. For example the <a href="https://panamapapers.icij.org/">Panama Papers</a> (2016) and the <a href="http://www.bbc.com/news/world-41880153">Paradise Papers</a> (2017). 

The Paradise Papers is a cache of some 13GB of data that contains 13.4 million confidential records of offshore investment by 120,000 people and companies in 19 tax jurisdictions. These numbers shows that the leak is very large and probably has a large impact on taxes. 

It was published by the International Consortium of Investigative Journalists (ICIJ) on November 5, 2017. This organization made a nice informative video about the leak. This video can be viewed <a href="https://www.icij.org/investigations/paradise-papers/watch-the-paradise-papers-secrets-of-the-global-elite/">here</a>. 

A lot of famous and policitally important people and involved in the Papers. For example Juan Manual Santos (the President of Columbia), Wilbur Ross (U.S. Secretary of Commerce) and Queen Elizabeth II. Besides, also large companies like Facebook, Nike and Yahoo are mentioned in the leak.

However, it is not only interesting to see which persons and which firms are mentioned in the papers, it is also very interesting to check which countries are mainly involved in the published papers and take as a result the most advantage of the existence of so-called tax havens. Besides, it might be interesting to see when the specific countries set up a lot of companies in the tax havens or at which time the number of shawm companies related to specific countries declined.

# Conclusion

It follows from this small study that Hong Kong, Switzerland, British Virgin Islands and to a lower extent Panama, Jersey, Luxembourg and the United Kingdom are mentioned most often in the published Papers. As a result, it seems that these countries got the largest benefits due to the existence of tax havens.

On the other hand, companies from countries like Switzerland, Hong Kong, the United Kingdom and Jersey are often not active anymore. Sometimes these are dissolved or defaulted. The reason for this fact is difficult to find. Maybe it became more expensive to set up shawn companies. Or costs for hiding the tax systems became higher. It can also be for countries like the United Kingdom and Switzerland that the European Union (or other organizations) played a role in the decline of entities. Often, organizations like the European Union, wants to increase tax equality and demotivate investments in tax havens. So it can be the case that under pressure of these organizations countries like Switserland inactivated their entities in tax havens.
